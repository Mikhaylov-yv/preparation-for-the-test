# Нахождение точек пересечения с кривой пстороенной по точкам

```python
import numpy as np
from scipy.interpolate import UnivariateSpline

import matplotlib.pyplot as plt
```

## Строим исходную скривую

```python
x = np.array([0, 0.2, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1])
X = np.reshape(x, (9, 1))
y = np.array([0, 10, 20, 30, 40, 50, 60, 80, 100])
```

```python
plt.plot(X, y, marker = 'o')
plt.show()
```

## Функция построения графика

```python
def pred_line(X, y, model):
    fig = plt.figure(figsize=(16, 16))
    plt.grid()
    plt.plot(X, y, marker = 'o')
    X = np.arange(X.min(), X.max(), 0.02)
    X = np.reshape(X, (len(X), 1))
    try:
        y_pred = model.predict(X)
    except:
        y_pred = model(X)
    plt.plot(X, y_pred, marker = 'x')
    plt.show()
```

```python
spl = UnivariateSpline(x, y,
                       s = 1
                      )
er = spl(x) - x
# print(er)
er[er < 0] = er[er < 0] * -1
print(er.sum())
pred_line(X, y, spl)
```
